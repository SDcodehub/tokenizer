```python
from peft import AutoPeftModelForCausalLM
from transformers import AutoTokenizer


max_seq_length = 2048
dtype = None


model = AutoPeftModelForCausalLM.from_pretrained(
    "Telugu-LLM-Labs/Telugu-gemma-7b-finetuned-sft",
    load_in_4bit = False,
    token = hf_token
)
tokenizer = AutoTokenizer.from_pretrained("Telugu-LLM-Labs/Telugu-gemma-7b-finetuned-sft")
```

```python
alpaca_prompt = """Below is an instruction that describes a task paired with an input that provides the context. Write a response that correctly completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""
```

```python
inst = "ఏనుగు మీద ఒక పిల్లల కథ రాయండి"

inputs = tokenizer(
[
    alpaca_prompt.format(
        inst, # instruction
        "", # input
        "", # output - leave this blank for generation!
    )
], return_tensors = "pt").to("cuda")
model = model.to("cuda")

outputs = model.generate(**inputs, max_new_tokens = 400)
result = tokenizer.batch_decode(outputs)[0]
print (result)
```

```python
from transformers import AutoTokenizer
from collections import OrderedDict

def collect_tokens(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    vocab = tokenizer.get_vocab()

    # Sort the vocabulary by token ID
    sorted_vocab = sorted(vocab.items(), key=lambda x: x[1])

    # First 4 tokens
    first_4_tokens = list(sorted_vocab)[:4]

    # Tokens that match 'positive', 'negative', or 'neutral' when lowercased
    sentiment_tokens = []
    for token, token_id in vocab.items():
        if token.lower() in ['positive', 'negative', 'neutral']:
            sentiment_tokens.append((token, token_id))

    # Combine tokens and remove duplicates while preserving order
    all_tokens = first_4_tokens + sentiment_tokens
    unique_tokens = list(OrderedDict.fromkeys(all_tokens))

    # Count total tokens
    total_tokens = len(unique_tokens)

    return total_tokens, unique_tokens

total_count, tokens_and_ids = collect_tokens("Telugu-LLM-Labs/Telugu-gemma-7b-finetuned-sft")

print(f"Total number of tokens: {total_count}")

print("Tokens and their token IDs:")
for token, token_id in tokens_and_ids:
    print(f"Token: {token}, Token ID: {token_id}")
```

```python
import torch

# Create a mapping between the original token IDs and the new token IDs
token_id_mapping = {token_id: new_token_id for new_token_id, (token, token_id) in enumerate(tokens_and_ids)}

print (token_id_mapping)
```

```python
# Get the last embedding decoder layer
last_embedding_decoder_layer = model.base_model.model.lm_head

# Create a new last embedding decoder layer with the subset of tokens_and_ids
new_last_embedding_decoder_layer = torch.nn.Linear(last_embedding_decoder_layer.in_features, len(tokens_and_ids))

# Copy the weights of the subset of tokens_and_ids from the original last embedding decoder layer
for token, token_id in tokens_and_ids:
    new_token_id = token_id_mapping[token_id]
    new_last_embedding_decoder_layer.weight.data[new_token_id] = last_embedding_decoder_layer.weight.data[token_id]

# Set the new last embedding decoder layer in the model
model.base_model.model.lm_head = new_last_embedding_decoder_layer
```

```python
# Create a reverse mapping from new token IDs to their corresponding tokens
reverse_token_id_mapping = {new_token_id: token for new_token_id, (token, _) in enumerate(tokens_and_ids)}

print (reverse_token_id_mapping)
```

## Some explanation

In the original model, the last embedding decoder layer (`model.base_model.model.lm_head`) is a linear layer that projects the hidden states from the model's dimensions to the vocabulary size. The dimensions of this layer are typically `(hidden_size, vocab_size)`, where `hidden_size` is the size of the hidden states in the model, and `vocab_size` is the total number of tokens in the model's vocabulary.

For example, if the model has a hidden size of 768 and a vocabulary size of 50,000, the dimensions of the last embedding decoder layer would be `(768, 50,000)`. This means that for each token in the vocabulary, there is a corresponding embedding vector of size 768.

Now, when we want to trim the model to use only a subset of tokens, we can create a new last embedding decoder layer with reduced dimensions. In the code snippet you provided, we create a new linear layer (`new_last_embedding_decoder_layer`) with the same input features as the original layer (`last_embedding_decoder_layer.in_features`), but the output features are set to the length of the subset of tokens we want to keep (`len(tokens_and_ids)`).

So, if we have a subset of 12 tokens that we want to use, the dimensions of the new last embedding decoder layer would be `(768, 12)`. This means that we are effectively reducing the vocabulary size from 50,000 to just 12 tokens.

To ensure that the new last embedding decoder layer has the correct weights for the subset of tokens, we copy the weights from the original layer to the new layer. We iterate over the `tokens_and_ids` list, which contains the token-token ID pairs for the subset of tokens we want to keep. For each token, we map its original token ID to the new token ID using the `token_id_mapping` dictionary. Then, we copy the corresponding weight vector from the original layer to the new layer at the new token ID position.

By doing this, we are effectively extracting the relevant embedding vectors for the subset of tokens and discarding the rest. The new last embedding decoder layer will only output probabilities for the subset of tokens we specified, reducing the computational overhead and focusing the model on the specific tokens of interest.

Finally, we replace the original last embedding decoder layer in the model with the new layer using `model.base_model.model.lm_head = new_last_embedding_decoder_layer`. This ensures that the model uses the trimmed embedding layer during inference.

We also print the dimensions of the new last embedding decoder layer and the total number of parameters in that layer. This helps verify that the layer has been successfully trimmed to the desired subset of tokens.

In summary, by creating a new last embedding decoder layer with reduced dimensions and copying the weights of the subset of tokens from the original layer, we are able to trim the model to focus only on the specific tokens of interest, thereby reducing the computational overhead and adapting the model to our specific use case.
